```python
from pymongo import MongoClient
import pandas as pd
```

```python
def connectCollection(database, collection):
    client = MongoClient()
    db = client[database]
    coll = db[collection]
    return db, coll
```

```python
db, coll = connectCollection('companies','companies')
```
