## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
df = pd.read_csv('../weather-raw.csv')


del df['year']
del df['id']



display(df)
```

```python
dfmax= df[df['element']=="tmax"]
dfmin= df[df['element']=="tmin"]

dfmax.head(10)
dfmin.head(10)

```

```python
dfmin['Averago'] = dfmin.mean(numeric_only=True, axis=1)
dfmin['Mon'] = dfmin.min(numeric_only=True, axis=1)
dfmin['Mox'] = dfmin.max(numeric_only=True, axis=1)

dfmax['Average'] = dfmax.mean(numeric_only=True, axis=1)
dfmax['Min'] = dfmax.min(numeric_only=True, axis=1)
dfmax['Max'] = dfmax.max(numeric_only=True, axis=1)
```

```python
dsemifinal = dfmax.merge(dfmin, on="month")
display(dsemifinal)
```

```python
import calendar
dsemifinal['month'] = dsemifinal['month'].apply(lambda x: calendar.month_abbr[x])
display(dsemifinal)
```

```python
dfinal=dsemifinal[['month','Average','Min','Max','Averago','Mon','Mox',]]
dfinal=dfinal.rename(columns={ 'Average': "Average Tmin", 'Min': "Min Tmin", 'Max':'Max Tmin', 'Averago':'Average Tmax', 'Mon':'Min Tmax', 'Mox':'Max Tmax'})

dfinal= dfinal.round(2)
```
